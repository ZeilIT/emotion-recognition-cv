## Импорт модулей и вспомогательные функции

```python
import cv2
import numpy as np
from deepface import DeepFace
from google.colab import files
from matplotlib import pyplot as plt

def detect_emotions(image_path):
    """Анализирует изображение, возвращает список лиц с эмоциями и координатами."""
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Не удалось загрузить: {image_path}")
    try:
        results = DeepFace.analyze(img_path=image_path,
                                   actions=['emotion'],
                                   enforce_detection=False,
                                   detector_backend='opencv')
    except Exception as e:
        print(f"Ошибка анализа: {e}")
        return []
    if not isinstance(results, list):
        results = [results]
    faces_info = []
    for face in results:
        region = face['region']
        x, y, w, h = region['x'], region['y'], region['w'], region['h']
        emotion = face['dominant_emotion']
        confidence = face['emotion'][emotion]
        faces_info.append({
            'bbox': (x, y, w, h),
            'emotion': emotion,
            'confidence': confidence
        })
    return faces_info

def draw_result(image_path, faces_info):
    """Рисует рамки и подписи на изображении, возвращает картинку RGB."""
    img = cv2.imread(image_path)
    for face in faces_info:
        x, y, w, h = face['bbox']
        emotion = face['emotion']
        conf = face['confidence']
        cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
        label = f"{emotion} ({conf:.1f}%)"
        cv2.putText(img, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img_rgb

print("Библиотеки импортированы, функции готовы.")
```

## Распознавание эмоций
Запустите ячейку ниже - она покажет эмоцию и обведёт лицо.

```python
if image_path:
    faces = detect_emotions(image_path)
    if not faces:
        print("Лица не обнаружены. Попробуйте другое фото (лицо анфас, хорошее освещение).")
    else:
        print(f"\nНайдено лиц: {len(faces)}")
        for i, face in enumerate(faces, 1):
            x, y, w, h = face['bbox']
            print(f"  Лицо {i}: координаты ({x},{y},{w},{h}) — эмоция {face['emotion']} ({face['confidence']:.1f}%)")
        img_rgb = draw_result(image_path, faces)
        plt.figure(figsize=(8,6))
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.title(f"Результат: {len(faces)} лиц(о)")
        plt.show()
else:
    print("Сначала загрузите изображение в ячейке 6.")
```
